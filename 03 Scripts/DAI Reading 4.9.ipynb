```python
#creating list holding true/false values to the test np.random.rand() <= 0.7

np.random.seed(4)
dev = np.random.rand(len(df_opm)) <= 0.7
```

```python
#store 70% of sample in dataframe big

big = df_opm[dev]
```

```python
#store 30% of sampe in dataframe small

small = df_opm[~dev]
```

```python
len(df_opm)
```
